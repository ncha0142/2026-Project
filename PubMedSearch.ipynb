A folder of abstract texts in .txt format was then generated based on the original paper.

```python
#take a list of PMIDs
#loop over all of them
#write their abstract contents into txt files
#Path to PMID list: 
import urllib.request
import os
import time

PMID_list = r"C:\Users\noahj\Desktop\Prof Phan Research\2026 Project\Assets\PMIDtotal.txt"

os.chdir(r"C:\Users\noahj\Desktop\Prof Phan Research\2026 Project\Assets\Abstracts_All")

counter = 1
with open(PMID_list, "r") as file:
    for line in file:
        #keep track
        #print("PMID: ", line, "  ", counter)
        counter += 1
        
        try:
            time.sleep(0.5)
            url = "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi?db=pubmed&id={0}&retmode=text&rettype=abstract".format(line.strip("\n"))
            abst = urllib.request.urlopen(url)
        
            with open("{}.txt".format(line.strip()), "w") as output:
                for line in abst:
                    decoded_line = line.decode("utf-8")
                    output.write(decoded_line)
                output.close()

            print(counter)

        except:
            pass
```
